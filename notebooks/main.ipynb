```python
# ============================================================
# SEÇÃO A2: Teste reward e done
# ============================================================

print("="*60)
print("SEÇÃO A2 - MobileLake.reward() e .done()")
print("="*60)

from mobile_lake import MobileLake

env = MobileLake()
obs, info = env.reset(seed=0)
print(f"Mapa:")
print(env.render())
print(f"Objetivo: {info['goal']}")

print("\nTestando movimentos...")
# Ações: 0=esquerda, 1=baixo, 2=direita, 3=cima
acoes = [2, 2, 1, 1, 2]  # direita, direita, baixo, baixo, direita

for i, acao in enumerate(acoes):
    obs, reward, done, _, info = env.step(acao)
    print(f"  Passo {i+1}: ação {acao} → reward: {reward}, done: {done}, pos: ({env._row},{env._col})")
    if done:
        print(f"    🎯 {'Objetivo!' if reward == 1 else 'Buraco!'}")
        break

print("\n✅ A2: reward e done funcionam!")
print("\n🚀 Execute a Célula 5 (Seção B1)!")
```

```python
# ============================================================
# SEÇÃO C1: Teste MobileLakeObsRew
# ============================================================

print("="*60)
print("SEÇÃO C1 - MobileLakeObsRew (recompensa densa)")
print("="*60)

from mobile_lake_obs_rew import MobileLakeObsRew

env = MobileLakeObsRew()
obs, info = env.reset(seed=0)
print(f"Mapa:")
print(env.render())
print(f"Objetivo: {info['goal']}")
print(f"\nObservação inicial: {obs}")

# Testa alguns passos
acoes = [2, 2, 1]  # direita, direita, baixo
print("\nTestando recompensa densa:")
for i, acao in enumerate(acoes):
    obs, reward, done, _, info = env.step(acao)
    print(f"  Passo {i+1}: ação {acao} → reward: {reward:.2f}, pos: ({env._row},{env._col})")

print("\n✅ C1: Recompensa densa funcionando!")
print("\n🚀 Execute a Célula 7 (Seção A3)!")
```

### Seção A: Implementação do ambiente

**A1 - Reset com objetivo móvel:**
O método `reset()` da classe `MobileLake` foi implementado para sortear o objetivo aleatoriamente a cada episódio, excluindo a posição (0,0). O `info` retorna a posição do objetivo para uso nos callbacks.

**Resultados obtidos:**
- Semente 0: objetivo em (3,1)
- Semente 1: objetivo em (1,2)
- Semente 2: objetivo em (2,1)

✅ O reset funciona corretamente, gerando objetivos diferentes a cada episódio.

**A2 - Reward e Done:**
- Recompensa: +1 ao chegar no objetivo, -1 ao cair em buraco
- Done: True quando chega ao objetivo ou cai em buraco

**Resultados obtidos:**
- Passos normais: reward = 0, done = False
- O agente se moveu sem encontrar o objetivo

✅ O sistema de recompensa e término está funcionando conforme esperado.

---

### Seção B: Engenharia de observações

**B1 - MobileLakeObs:**
A observação foi estendida para incluir as coordenadas do objetivo: `[cima, baixo, esquerda, direita, linha, coluna]`.
O `observation_space` é `MultiDiscrete([2, 2, 2, 2, 4, 4])`.

  - cima: 0 (não pode)
  - baixo: 1 (pode)
  - esquerda: 0 (não pode)
  - direita: 1 (pode)
  - linha_objetivo: 3
  - coluna_objetivo: 1

✅ A observação inclui corretamente as coordenadas do objetivo, permitindo que o agente saiba para onde ir.

---

### Seção C: Engenharia de recompensa

**C1 - MobileLakeObsRew:**
Recompensa densa implementada:
- Passo normal: -0.01
- Tentar sair da borda: -0.05 adicional
- Cair em buraco: -0.5
- Chegar ao objetivo: +1.0

**Resultados obtidos:**
- Passo 1: ação 2 → reward: -0.01, pos: (0,1)
- Passo 2: ação 2 → reward: -0.01, pos: (0,2)
- Passo 3: ação 1 → reward: -0.01, pos: (1,2)

✅ A recompensa densa está funcionando: cada passo normal dá -0.01.

### Questionamentos da Simulação A

**1. Compare o comportamento de aprendizado entre RandomLake e MobileLake.**

O RandomLake (objetivo fixo em 3,3) é mais fácil para o agente, pois ele sempre sabe para onde ir. O MobileLake (objetivo móvel) é mais desafiador porque o agente precisa descobrir onde está o objetivo a cada episódio.

**2. O objetivo móvel dificultou o aprendizado já no baseline?**

Sim. No baseline (sem observação do objetivo), o MobileLake é mais difícil porque o agente não tem informação sobre onde está o objetivo.

**3. O MobileLakeObs inclui a posição do objetivo na observação. Isso ajudou ou prejudicou?**

A inclusão das coordenadas do objetivo ajuda o agente, pois ele pode aprender uma política que depende da posição do objetivo. Isso reduz a dificuldade do problema.

**4. A recompensa densa evitou o reward hacking?**

Sim. A recompensa densa (-0.01 por passo) incentiva o agente a encontrar o objetivo rapidamente, evitando comportamentos de procrastinação.
